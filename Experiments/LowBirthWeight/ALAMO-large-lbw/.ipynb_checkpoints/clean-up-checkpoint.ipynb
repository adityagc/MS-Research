```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("/home/adityagc/Desktop/MS-Research/Experiments/LowBirthWeight/ALAMO-large-lbw/datagrpreg.csv")
```

```python
df1.head()
```

```python
bwt = df1['bwt']
del df1['bwt']
del df1['low']
del df1['Unnamed: 0']
df1['bwt'] = bwt
df1.head()
```

```python
df1.dtypes
```

```python
df1['white'] = df1['white'].astype(int)
df1['black'] = df1['black'].astype(int)
df1['smoke'] = df1['smoke'].astype(int)
df1['ftv1'] = df1['ftv1'].astype(int)
df1['ftv2'] = df1['ftv2'].astype(int)
df1['ftv3m'] = df1['ftv3m'].astype(int)
df1['ptl1'] = df1['ptl1'].astype(int)
df1['ptl2m'] = df1['ptl2m'].astype(int)
# df1.dtypes
df1.head()
```

```python
maxvals = df1.max()
maxarr = maxvals.values
np.savetxt("maxarr.csv", maxarr, newline=" ")
```

```python
minvals = df1.min()
minarr = minvals.values
np.savetxt("minarr.csv", minarr, newline=" ")
```

```python
df1.to_csv('cleaned.csv', sep="\t", index=False)
```
